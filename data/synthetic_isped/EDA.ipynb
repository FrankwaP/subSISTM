```r
library(ggplot2)

# install.packages('SmartEDA')
library(SmartEDA)  # https://daya6489.github.io/SmartEDA/index.html
```

```r
data <- read.csv2('simulation.csv')
data
```

```r
data$individus <- as.factor(data$individus)
```

```r
N_max <- 10
```

```r
x_val = "x1"

data_temp <- subset(data, select=c('individus', 'temps', x_val), subset=as.numeric(data$individus)<=N_max) 
ggplot(data_temp, aes(x=temps, .data[[x_val]], group=individus, color=individus)) + geom_line()
```
